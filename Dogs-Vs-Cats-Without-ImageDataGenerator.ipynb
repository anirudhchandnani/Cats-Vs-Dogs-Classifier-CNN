```python
import numpy as np
import pandas as pd 
from keras.preprocessing.image import ImageDataGenerator, load_img
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import random
import os
import zipfile
#print(os.listdir("../input"))
```

```python
def extract_files(source_path, target_path):
    zip_ref = zipfile.ZipFile(source_path,'r')
    zip_ref.extractall(target_path)
    zip_ref.close()
```

```python
import cv2
main_dir = "desktop/train/"
train_dir = "train"
path = os.path.join(main_dir,train_dir)
X = []
Y = []
for img in os.listdir("desktop/train/train"):
    category = img.split(".")[0]
    img_array = cv2.imread(os.path.join(path,img))
    new_img_array = cv2.resize(img_array, dsize=(150, 150))
    #image = load_img(os.path.join(path,img))
    #plt.figure()
    #plt.imshow(image)
    X.append(new_img_array)
    Y.append(category)
```

```python
X= np.array(X)
X.shape
```

```python
Y=np.array(Y)
Y.shape
```

```python
y_df = pd.DataFrame(Y)
y_df[0]=y_df[0].replace({"dog":1 , "cat":0})
Y = np.array(y_df[0])
Y
```

```python
X= X/255
```

```python
X
```

```python
X = X.reshape(25000, 150,150,3)
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten, Dense, Activation, BatchNormalization
model = Sequential()

# 3 Convolutional, Pooling, Dropout layers
model.add(Conv2D(8, 
                 (3,3), 
                 activation = 'relu', 
                 input_shape = (150, 150, 3)))
model.add(MaxPooling2D(2,2))
model.add(Dropout(0.25))

model.add(Conv2D(16, 
                 (3, 3),
                 activation = 'relu'))
model.add(MaxPooling2D(2,2))
model.add(Dropout(0.25))

model.add(Conv2D(32, 
                 (3,3), 
                 activation = 'relu'))
model.add(MaxPooling2D(2,2))
model.add(Dropout(0.25))

# Dense layers start
model.add(Flatten())

model.add(Dense(1024, 
                activation = 'relu'))
model.add(Dropout(0.45))

model.add(Dense(256, 
                activation = 'tanh'))
model.add(Dropout(0.45))

model.add(Dense(1, 
                activation = 'sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss = 'binary_crossentropy', 
              optimizer = 'adam',
              metrics = ['accuracy'])
```

```python
model.fit(X, Y, epochs=2, batch_size=32, validation_split=0.2)
```
